# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
x= dataset.iloc[:, 1:-1].values
y =dataset.iloc[:, -1].values
```

# Trainig a linear model on the dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Training the Polynomial Regression model on the Training set

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree =4)
x_poly  = poly_reg.fit_transform(x)
regressor_2 = LinearRegression()
regressor_2.fit(x_poly, y)
```

#visualising the polynomial results

```
plt.scatter(x,y, color='red')
plt.plot(x,regressor_2.predict(poly_reg.fit_transform(x)),color ='blue')
plt.title('truth or bluff')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

#visualising linear results

```
plt.scatter(x,y, color='red')
plt.plot(x,lin_reg.predict(x),color ='blue')
plt.title('truth or bluff')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

# predicting new result with  linear

```
lin_reg.predict([[6.5]])
```

# predicting new result with polynomial

```
regressor_2.predict(poly_reg.fit_transform([[6.5]]))
```
